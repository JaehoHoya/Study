## GET 방식 요청

```python
import urllib.request
import urllib.parse
params = urllib.parse.urlencode({'name': '제임스', 'age': 10})
print("URL 인코딩 규칙이 적용된 문자열 : %s" % params)
url = "http://unico2013.dothome.co.kr/crawling/get.php?%s" % params    # Query String에 추가한다
print("URL 문자열 : %s" % url)
with urllib.request.urlopen(url) as f:
    print(f.read().decode('utf-8'))
```

## POST 방식 요청

```python
import urllib.request
import urllib.parse

data = urllib.parse.urlencode({'name': '제임스', 'age': 10})
print("URL 인코딩 규칙이 적용된 문자열 : %s" % data)
postdata = data.encode('ascii')

print("변환된 바이트 문자열 : %s" % postdata)
url = "http://unico2013.dothome.co.kr/crawling/post.php"

with urllib.request.urlopen(url, postdata) as f:          # POST 방식에서는 요청 Body에 담아 전송해야 한다
    print(f.read().decode('utf-8'))
```

```python
import urllib.request
import urllib.parse
data = urllib.parse.urlencode({'name': '스미스', 'age': 10})
postdata = data.encode('ascii')

req = urllib.request.Request(url='http://unico2013.dothome.co.kr/crawling/post.php',
            data=postdata)       # POST 방식 데이터를 전달하는 또 다른 방법
print(req)
with urllib.request.urlopen(req) as f:
    print(f.read().decode('utf-8'))
```

## requests.request()함수를 사용하여 웹페이지 소스 가져오기
* requests 모듈은 파이썬 표준 모듈은 아니고 Anaconda 에 포함되어 있음
* 설치 : conda install requests, 혹은 pip install requests
* urllib.request 가 GET, POST만 지원하는 반면, requests는 모든 요청이 가능하고  자동으로 인코딩도 지원함
* params : GET 방식 요청
* data : POST 방식 요청
* requests.request() : get, post 모두 가능
* requests.get() : get 요청
* requests.post() : post 요청
* Response.text : 문자열 응답
* Response.content : 바이너리 응답(문자열일 경우에는 문자열로 변환해야 함)

```python
import requests

r = requests.request('get', 'http://unico2013.dothome.co.kr/crawling/exam.html')
r.encoding = 'utf-8'
print(type(r))
if r.text :
    print(r.text)
else :
    print('응답된 콘텐츠가 없어요')
print('----------------------------------------------------------')

r = requests.request('head', 'http://unico2013.dothome.co.kr/crawling/exam.html')
r.encoding = 'utf-8'
print(type(r))
if r.text :
    print(r.text)
else :
    print('응답된 콘텐츠가 없어요')
print('----------------------------------------------------------')

r = requests.request('post', 'http://unico2013.dothome.co.kr/crawling/post.php', data= {'name':'백도', 'age' : 12})
r.encoding = 'utf-8'
print(type(r))
if r.text :
    print(r.text)
else :
    print('응답된 콘텐츠가 없어요')
```

## 태그 이름을 사용하여 다수개의 태그 가져오기
* find_all()
* find()
* select() : CSS Selector를 사용하여 태그 선택하기, select('div'), select('#div1'), select('.cls1'), select('a>b'), select('a b'), select('img[src=a.png]'), select('img[src=png]')

```python
html_doc = """
<!DOCTYPE html>
<html>
     <head>
          <meta charset='utf-8'>
          <title>Test BeautifulSoup</title>
     </head>
     <body>
          <p align="center">P태그의 컨텐트</p>
          <img src="http://unico2013.dothome.co.kr/image/flower.jpg" width="300">
          <ul>
               <li>테스트1<strong>강조</strong></li>
               <li>테스트2</li>
               <li>테스트3</li>
          </ul>
     </body>
</html> """

from bs4 import BeautifulSoup
bs = BeautifulSoup(html_doc, 'html.parser')

print("[[[ string 속성 ]]]")
print('1.', type(bs.p.string), ':', bs.p.string)     # <class 'bs4.element.NavigableString'>
print('2.', type(bs.ul.string), ':', bs.ul.string)
print('3.', type(bs.ul.li.string), ':', bs.ul.li.string)
print('4.', type(bs.ul.li.strong.string), ':', bs.ul.li.strong.string)
print("[[[ text 속성 ]]]")
print('5.', type(bs.p.text), ':', bs.p.text)         # <class 'str'>
print('6.', type(bs.ul.text), ':', bs.ul.text)
print('7.', type(bs.ul.li.text), ':', bs.ul.li.text)
print('8.', type(bs.ul.li.strong.text), ':', bs.ul.li.strong.text)
print("[[[ contents 속성 ]]]")
print('9.', type(bs.p.contents), ':', bs.p.contents)
print('10.', type(bs.ul.contents), ':', bs.ul.contents)
print('11.', type(bs.ul.li.contents), ':', bs.ul.li.contents)
print('12.', type(bs.ul.li.strong.contents), ':', bs.ul.li.strong.contents)
print("[[[ find_all()의 text=True 매개변수 ]]]")
print(bs.find_all("li", text=True))  # 콘텐츠가 텍스트로만 구성된
```

```python
html_doc="""
<!DOCTYPE html>
<html>
  <head>
     <meta charset='utf-8'>
     <title>Test BeautifulSoup</title>
  </head>
  <body>
     <p align="center"> text contents </p>
     <p align="right">  text contents 2 </p>
     <p align="left">   text contents 3 </p>
     <img src="http://unico2013.dothome.co.kr/image/flower.jpg" width="500">
     <div>
       <p>text contents 4</p>
     </div>
  </body>
</html> """

from bs4 import BeautifulSoup
bs = BeautifulSoup(html_doc, 'html.parser')

print('1.', type(bs.find('p')))                    # 못찾으면 None 리턴
print('2.', type(bs.find_all('p')))
print("---------------------------------")
print('3.', bs.find('title'))
print('4,', bs.find('p'))
print('5.', bs.find('img'))
print("---------------------------------")
ptags = bs.find_all('p')                     # ResultSet 객체 리턴
print('6.',ptags)
print("----------------")
for tag in ptags :
    print(tag)
```

```python
import requests
from bs4 import BeautifulSoup
import re

req = requests.get('http://movie.naver.com/movie/point/af/list.nhn?page=1')
html = req.text
soup = BeautifulSoup(html, 'html.parser')
titles = soup.select('.movie')
points = soup.select('td.title > div > em')
reviews = soup.select('td.title')

movie_title = []
movie_point = []
movie_review = []

for dom in titles:
    movie_title.append(dom.text)
for dom in points:
    movie_point.append(dom.text)
for dom in reviews:
    content = dom.contents[6]
    content=re.sub("[\n\t]", '', content)
    movie_review.append(content)
commentLength = len(movie_title)
for i in range(commentLength):
    print("영화 제목 : " + movie_title[i])
    print("평점 : " + movie_point[i])
    print("리뷰글 : " + movie_review[i])
    print("-----------------------------------------")
```

```python
import requests
from bs4 import BeautifulSoup
import re
for n in range(1,31):
    req = requests.get('http://movie.naver.com/movie/point/af/list.nhn?page='+str(n))
    html = req.text

    soup = BeautifulSoup(html, 'html.parser')
    titles = soup.select('.movie' )
    points = soup.select('td.title > div > em')
    reviews = soup.select('td.title')

    movie_title = []
    movie_point = []
    movie_review = []
    for dom in titles:
        movie_title.append(dom.text)
    for dom in points:
        movie_point.append(dom.text)
    for dom in reviews:
        content = dom.contents[6]
        content=re.sub("[\n\t]", '', content)
        movie_review.append(content)

    commentLength = len(movie_title)
    for i in range(commentLength):
        print(movie_point[i] + ","+movie_title[i]+","+movie_review[i])
    print("-----------------------------------------------------")
```

```python
import requests
from bs4 import BeautifulSoup
title = []
link = []
urlstr = 'http://www.yes24.com/SearchCorner/Search?domain=BOOK&query=python'
r = requests.get(urlstr)
r.encoding = "utf-8"
bs = BeautifulSoup(r.text, 'html.parser')
bookList = bs.select('a.gd_name')    # 클래스로 찾기
# 크롬 > 도구 더보기 > 개발자 도구 > Inspect 아이콘 클릭 > 웹 페이지에서 특정 내용 드래그 선택 후 inspect 화면에서 해당 태그 찾기


for bookDom in bookList:
    title.append(bookDom.string)
    link.append(bookDom["href"])

print("-- 도서 제목 --")
print(title)
print("-- 도서 링크 URL --")
print(link)

# 제목과 링크를 csv 파일에 저장하는 예
with open('/content/drive/MyDrive/Python_Basic/booklink.csv', "wt") as f:
    f.write('booktitle,booklink\n')
    for i in range(len(title)):
        f.write(title[i]+","+link[i]+'\n')
```

## 네이버 만화 사이트 페이지 이동하면서 데이터 가져오기
* https://comic.naver.com/bestChallenge/list?titleId=695997&page=2&sort=DESC
* 제목
* 요약
* 관심점수

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


titles = list()
pub = list()
price = list()
rating = list()

for page_num in range(1,11):
    url = f"https://www.yes24.com/Product/Search?domain=ALL&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&page={page_num}"

    req = requests.get(url)

    html = req.content.decode("utf-8")
    bs= BeautifulSoup(html, "html.parser")

    title = bs.select("#yesSchList > li:nth-child(1) > div > div.item_info > div.info_row.info_name > a.gd_name")[0].text
    publisher = bs.select("#yesSchList > li:nth-child(1) > div > div.item_info > div.info_row.info_pubGrp > span.authPub.info_pub > a")[0].text
    pri = bs.select("#yesSchList > li:nth-child(1) > div > div.item_info > div.info_row.info_price > strong > em")[0].text
    try:
        rate = bs.select("#yesSchList > li:nth-child(1) > div > div.item_info > div.info_row.info_rating > span.rating_grade > em")[0].text
    except:
        rate = "None"

    titles.append(title)
    pub.append(publisher)
    price.append(pri)
    rating.append(rate)

    print(f"{len(titles)}개 추출")

yes24_book_dict = { 'title':titles, 'publisher' : pub, 'price':price, 'rating': rating}

df = pd.DataFrame(yes24_book_dict)
display(df)

df.to_csv('/content/drive/MyDrive/Python_Basic/yes24_book.csv', encoding='utf-8-sig')
```
