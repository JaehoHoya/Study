```python
# 전화번호 찾기
def find_by_phone(phone):
  for emp in emps:
    if emp['phone']==phone:
      return emp
  return None

# 목록
def emp_list():
  for emp in emps:
    print(f'번호:{emp["no"]},이름:{emp["name"]},전번:{emp["phone"]}')

#수정
def update_phone(phone):
  for emp in emps:
    if emp['phone']==phone:
      new_phone=input('새 전화번호를 입력하세요')
      emp['phone']=new_phone
      print('수정 완료')
#삭제
def delete_member(no):
  for i,emp in enumerate(emps):
    if emp['no']==no:
      del emps[i]
      print('삭제 완료')
      break
  else:
    print('해당 사원이 없습니다')

#추가
def add_member():
  num   = (int)(input ('번호를 입력하세요'))
  name  = input('이름을 입력하세요')
  phone = (int)(input('전화번호를 입력하세요'))
  member={'no':num,'name':name,'phone':phone}
  emps.append(member)
  print('추가 완료')
```

```python
# 사원정보 CRUD
# 사원의 번호 이름, 전화번호가 담긴 목록
# 키보드에서 전화번호를 이용하여 검색이 가능하게 함
emps=[
  {"no": 1, "name": 'jaeho', "phone": '1234'},
  {"no": 2, "name": 'hoya',  "phone": '5678'}]
```

```python
# 사원정보 검색, 수정, 삭제 함수 정의하고 사용하기
# 목록(s), 추가(a) ,검색(f), 수정(u), 삭제(d) ,종료(x)

while(True):
  print('목록(s), 추가(a) ,검색(f), 수정(u), 삭제(d) ,종료(x)')
  menu=input('메뉴를 선택하세요')

  if menu=='x':
     print('프로그램을 종료합니다')
     break;
  if menu=='s':
    emp_list()
  if menu=='a':
    add_member()
  if menu=='f':
    phone=input('검색할 전화번호를 입력하세요')
    emp=find_by_phone(phone)
    if emp is None:
      print('검색된 사원이 없습니다')
    else:
      print(f'번호:{emp["no"]},이름:{emp["name"]},전번:{emp["phone"]}')
  if menu=='u':
    phone=input('수정할 전화번호를 입력하세요')
    update_phone(phone)
  if menu=='d':
    no=int(input('삭제할 사원의 번호를 입력하세요'))
    delete_member(no)


```

```python
class Student:
  school_name='Python School' # 메모리에 한번만 있어도 되뮤  static 변수


  def __init__(self,no,name,phone,major): # 인스턴스
    self.no=no
    self.name=name
    self.phone=phone
    self.major=major

  def __str__(self): # self 인스턴스 참조
    return f'{self.no}\t{self.name}\t{self.phone}\t{self.major}'

  def __eq__(self,other):
    return self.no==other.no

# 클래스 변수인 school_name의 값을 초기화할 필요가 있다면 클래스 메소드를 선언하여 클래스 변수의 값을 할당하면 된다

  @classmethod
  def change_school_name(cls,name): # cls  클래스가 로드된것과 객체가 로드된거랑 틀림 클레스 참조: cls: 클래스 참조
    cls.school_name=name
```

```python
# Vector2D 방향과 크기을 가진 물리량
# 벡터를 이용하여 물리현상을 계산을 하려면 물리현상을 반영할 수 있도록 가감승제 연상이 재정의 되어야 한다
```

```python
class Vector2D:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  # 파이썬은 같은 이름의 메소드를 다수개 만들 수 없다

  def __add__(self,other): # A + B     +하면  class안에 __add__가 돌아감
    return Vector2D(self.x+other.x,self.y+other.y)

  def __sub__(self,other): # A - B
    return Vector2D(self.x-other.x,self.y-other.y)

  def __str__(self) -> str:
    return f'({self.x},{self.y})'

  def __mul__(self, other):
      # 벡터와 스칼라의 곱 (A * k)
    if isinstance(other,Vector2D):
      return self.x *other.x + self.y*other.y
    elif isinstance(other,int):
      return Vector2D(self.x*other,self.y*other)



  def dot(self, other):  # 벡터 내적 (A • B)
      return self.x * other.x + self.y * other.y

  def length(self):
    return (self.x **2 +self.y **2) ** 0.5
```

```python
a=Vector2D(1,2)
b=Vector2D(3,4)
```

```python
c= a+b #__add__가 돌아감
print(c) # 각 성분끼리 더해졌네
```

```python
# 벡터의 스칼라 곱을 수행할 수 있는 연산자 오버로드 실습

v1 = Vector2D(2, 3)
v2 = v1 * 3

print(v2)  # 출력: (6, 9)
```

```python
# 벡터 내적
v3 = Vector2D(4, 5)
dot_product = v1.dot(v3)  # v1과 v3의 내적
print(dot_product)  # 출력: 23
```
