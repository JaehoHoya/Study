## 데이터 합치기
* merge()
* join()
* concat()

```python
from pandas import DataFrame, Series
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'], #행 3
                    'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
# 컬럼이 같은거 끼리 합치기
df_merge=pd.merge(df1, df2 ,on='key')
```

```python
df_merge
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})

df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
#c가 빠쪄 버림 outer join 필요
pd.merge(df1, df2, how='outer')
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                    'data1': range(6)})

df2 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                    'data2': range(5)})
```

```python
pd.merge(df1, df2, on='key', how='left')
```

```python
pd.merge(df1, df2, how='inner')
```

```python
left  = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1, 2, 3]})

right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on='key1') # 컬럼명이 동일할때 구분되게 하는 방법
```

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right')) # 이름 지정
```

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                      'value': range(6)})

right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True) # 왼쪽에서 key컬럼을쓰고 오른쪽은 인덱스
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                     index=['a', 'c', 'e'],
                     columns=['Ohio', 'Nevada'])
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=['b', 'c', 'd', 'e'],
                      columns=['Missouri', 'Alabama'])
```

```python
left2
```

```python
right2
```

```python
left2.join(right2, how='outer')
```

```python
s1 = pd.Series([0, 1],    index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6],    index=['f', 'g'])
```

```python
pd.concat([s1, s2, s3]) # 시리즈 합치기
```

```python
data=pd.DataFrame(np.random.randn(1000, 4)) #1000행 4열
data.describe() # 통계치 확인
#1000개의 기술통계를
# 평균이 0이고 표준편차가 1
# 표준편차를넘어간? 데이터 제거  온도가 40도인데 어떤온도는 100? AI 비정확
```

```python
data.head()
```

```python
col=data[2] # 컬럼 이름  col 천개
col[np.abs(col) > 3] # 3보다 큰값
# 돌리고 col에 커서하면  그래프나옴 -3, 3 극소수를 버리자 (이상치)
```

```python
(np.abs(data)>3).any(axis=1) # 어느한개라도 3보다 크면 True  컬럼의
```

```python
data[(np.abs(data)>3).any(axis=1)]
# 문제가 되는 행 가져오기
```

```python
#  np_sign(data): data가 양수이면 1, 0이면 0,음수이면 -1 리턴함

data[np.abs(data) > 3] = np.sign(data) * 3
data.describe()
```
