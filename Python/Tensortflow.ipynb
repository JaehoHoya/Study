```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
import numpy as np
```

```python
def get_y(x):
    return 4 * x + 7
```

```python
x = np.random.randint(-100, 100, 500)
print(x)
y = get_y(x)
y.shape

# -100에서 100 사이의 정수 중 500개를 랜덤하게 생성하여 x라는 변수에 저장합니다. 이로 인해 x는 500개의 정수로 이루어진 NumPy 배열이 됩니다.
```

```python
X= x.reshape(-1,1) # 500행 1열 , X는 500행 1열의 2차원 배열로 재구성됩니다.
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=0)
'''
est_size=.2: 전체 데이터 중에서 20%를 테스트 세트로 사용하겠다는 의미입니다. 나머지 80%는 훈련 세트로 사용됩니다.
random_state=0: 난수 생성의 시드를 설정하여 결과의 재현성을 보장합니다. 동일한 random_state 값을 사용하면 매번 동일한 방식으로 데이터가 분리됩니다.



훈련 세트는 모델이 학습하는 데 사용되며, 모델의 가중치와 파라미터를 조정합니다.
테스트 세트는 모델의 성능을 평가하는 데 사용되며, 모델이 실제로 새로운 데이터에 대해 얼마나 잘 작동하는지를 확인합니다.

'''
```

```python
# X : 문제 , y: 정답
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape

'''

입력 파라미터:

X: 입력 데이터(특성)입니다. 여기서는 2차원 배열로, 500개의 샘플과 1개의 특성을 가진 배열입니다.
y: 타겟 데이터(정답)입니다. 여기서는 get_y(x) 함수에 의해 생성된 출력 배열입니다.
test_size=.2: 전체 데이터 중에서 20%를 테스트 세트로 사용하겠다는 의미입니다. 나머지 80%는 훈련 세트로 사용됩니다.
random_state=0: 난수 생성의 시드를 설정하여 결과의 재현성을 보장합니다. 동일한 random_state 값을 사용하면 매번 동일한 방식으로 데이터가 분리됩니다.
결과:

X_train: 훈련 세트의 입력 데이터(80%).
X_test: 테스트 세트의 입력 데이터(20%).
y_train: 훈련 세트의 타겟 데이터(80%).
y_test: 테스트 세트의 타겟 데이터(20%).
'''
```

```python
reg_model = Sequential()
#input_dim=1은 우리가 이 신경망에 하나의 숫자(입력)만 줄 거라는 뜻
reg_model.add( Dense(1, input_dim=1)) # Input Layer ,Fully Connected Layer
reg_model.add( Dense(128)) # Hidden Layer  10: 뉴런의 수
reg_model.add( Dense(64))
reg_model.add( Dense(31))
reg_model.add( Dense(16))
reg_model.add( Dense(1))  # Output Layer
# 1개의 뉴런을 가진 출력 층이에요.
# 신경망을 만듬
```

```python
reg_model.summary() # 인공 신경망 요약
# Total params: 203 (812.00 B) 총 파라미터 가중치 +bias
#Trainable params: 203 (812.00 B) 학습을 위한 파라미터
#Non-trainable params: 0 (0.00 B)
# 가중치가 많을 수록 복잡한 문제를 풀 수있다
```

```python
#가중치(Weight): 입력값에 곱해지는 수
```

```python
reg_model.compile(loss='mse',optimizer="adam",metrics=['mse'])

#예측한 값과 실제 값의 차이를 계산해서 얼마나 잘 맞는지를 측정
#mse : 평균 제곱 오차
#옵티마이저: adam은 신경망을 학습할 때 가중치를 조정하는 방법이에요.
#평가 지표: metrics에 mse를 추가하는 것은 모델이 학습하는 동안 성능을 모니터링하기 위한 방법이에요.
#의미: 이렇게 설정하면 모델이 학습할 때마다 손실(MSE)을 기록하고, 학습이 얼마나 잘 되고 있는지를 볼 수 있어요.
'''
손실 함수 (Loss Function): 모델의 예측과 실제 값 간의 차이를 측정하는 함수입니다. 여기서 'mse'는 Mean Squared Error(평균 제곱 오차)를 의미합니다.

최적화 알고리즘 (Optimizer): 손실 함수를 최소화하기 위해 가중치를 조정하는 알고리즘입니다. 여기서는 Adam 옵티마이저를 사용하고 있습니다.
Adam은 경사 하강법을 기반으로 하며, 학습률을 자동으로 조정하여 빠르고 효율적인 학습을 지원

metrics=['mse']:평가지표 (Metrics): 모델의 성능을 평가하기 위해 사용할 지표입니다. 여기서 'mse'는 평균 제곱 오차를 의미하며, 훈련 및 검증 과정에서 모델의 성능을 모니터링하는 데 사용

'''
```

```python
hist=reg_model.fit(X_train,y_train,epochs=30,batch_size=10,validation_split=.1)

#epochs=100 : 학습의 횟수
#batch_size=10 : 데이터 10개를 전달하고 에러의 평균을 내서 가중치를 갱신하겠다는 의미
#이 설정은 전체 데이터의 **10%**를 검증 데이터로 사용하겠다는 뜻이에요. 즉, 훈련 데이터에서 90%는 모델을 학습하는
#왜 필요한가?:
'''
모델을 학습하는 동안 항상 같은 데이터를 사용하면, 모델이 그 데이터에 너무 익숙해져서 새로운 데이터에 대한 성능이 떨어질 수 있어요. 이를 **과적합(Overfitting)**이라고 해요.
검증 데이터를 사용하면 모델이 학습 중에 보지 못한 데이터로 성능을 평가할 수 있어요. 이렇게 하면 모델이 잘 학습하고 있는지를 확인할 수 있죠.

하이퍼파라미터의 특징

모델의 매개 변수를 추정하는 데 도움이 되는 프로세스에서 사용된다.
하이퍼파라미터는 개발자에 의해 수동으로 설정할 수 있다.(임의 조정 가능)
학습 알고리즘의 샘플에 대한 일반화를 위해 조절된다.

하이퍼파라미터의 예

학습률
손실 함수
일반화 파라미터
미니배치 크기
에포크 수
가중치 초기화
은닉층의 개수
k-NN의 k값

'''
```

```python
hist.history.keys() # 학습 내역의 손실,손실^2의 평균 ,
```

```python
loss = hist.history['loss']
mse = hist.history['mse']
vloss = hist.history['val_loss']
vmae = hist.history['val_mse']
'''
검증 손실(val_loss):

vloss = hist.history['val_loss']는 검증 데이터에서의 손실 값을 가져오는 코드예요.
의미: 손실 값은 모델이 예측한 값과 실제 값의 차이를 나타내며, 낮을수록 모델의 예측이 실제 값에 가깝다는 뜻이에요.
이렇게 가져온 검증 손실 값을 사용하여 모델의 성능을 평가할 수 있어요.
검증 평균 제곱 오차(val_mse):

vmae = hist.history['val_mse']는 검증 데이터에서의 평균 제곱 오차(MSE)를 가져오는 코드예요.
의미: MSE는 예측 값과 실제 값의 차이를 제곱하여 평균을 낸 값으로, 손실과 유사하지만 다른 형태의 지표예요. 역시 낮을수록 좋죠.

'''
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,2))
plt.plot(loss, 'r--', label='loss')  # r -- 빨간 -- 선
# plot : Line(선) 차트
plt.plot(mse, 'g', label='mse')
plt.plot(vloss, 'y--', label='val_loss')
plt.plot(vmae, 'b', label='val_mse')
plt.legend()
plt.show()
```

```python
# y= 2x +5
# 1~10 사이의  테이터를 1000개 생성 x= np.linspace(1,10,1000)
# 노이즈 : np.random.normal(3,3.1000)
# y'= 2x+5 +noise  : 노이즈가 섞인 y 데이터
# Linear Regression , Tensorflow
# x= 15일때 의 y는?  35
```

```python
x = np.linspace(1,10,1000)
print(x)
```

```python
noise = np.random.normal(3,3,1000)
print(noise)
```

```python
def get_y2(x):
    return 2 * x + 5
```

```python
x = np.linspace(1,10,1000)
noise = np.random.normal(3,3,1000) #평균이 3, 표준편차가 3인 노이즈 생성 정규 분포
y=get_y2(x)+ noise
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y ,s=1) # s 점의 크기
plt.show()
# 아래 그래프를 관통하는 직선을 가져오기
```

```python
x.shape
```

```python
X= x.reshape(-1,1)
X.shape ,y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
reg_model = Sequential()
reg_model.add( Dense(1, input_dim=1))
reg_model.add( Dense(20))
reg_model.add( Dense(1 ,activation='linear'))  #활성 함수
```

```python
reg_model.summary()
```

```python
reg_model.compile(loss='mse',optimizer="adam",metrics=['mse'])
```

```python
hist=reg_model.fit(X_train,y_train,epochs=100,batch_size=10,validation_split=.1)
```

```python
loss = hist.history['loss']
mse = hist.history['mse']
vloss = hist.history['val_loss']
vmae = hist.history['val_mse']
'''
lose= 훈련 데이터에 대한 손실 값입니다. 모델이 훈련 과정에서 얼마나 잘 학습했는지를 나타냅니다.

해석:lose, mes(평균제곱 오차),이 값이 작을수록 모델이 훈련 데이터에 잘 적합하고 있다는 것을 의미합니다. 즉, 예측값과 실제값 간의 차이가 적다는 것입니다.

vloss =검증 데이터에 대한 손실 값입니다. 모델이 검증 데이터에 대해 얼마나 잘 예측하는지를 평가합니다.
해석: vloss,vmae이 값이 낮을수록 모델이 검증 데이터에 대해 잘 적합하고 있다는 것을 의미합니다. 훈련 데이터에 비해 검증 데이터의 손실 값이 높으면 과적합(overfitting) 가능성을 나타낼 수 있습니다.



'''
```

```python
plt.figure(figsize=(5,2))
plt.plot(loss, 'r--', label='loss')  # r -- 빨간 -- 선
# plot : Line(선) 차트
plt.plot(mse, 'g', label='mse')
plt.plot(vloss, 'y--', label='val_loss')
plt.plot(vmae, 'b', label='val_mse')
plt.legend()
plt.show()
```

```python
reg_model.evaluate(X_test,y_test)
```

## 다중변수 함수의 값 회귀 분석
* y= 3x1+ 5x2 + 7 + noise

* x1: 500 개
* x2 :500 개
* x1.y 산점도 표시
* x2. y 산점도 표시
* input_dim =2
X.shape -> (500,2)
[[x1,x2],[x1,x2]...]

```python
x1=np.random.randint(1,100,500)
x2=np.random.randint(1,100,500)
```

```python
y=3 * x1 + 5 * x2 +7 + np.random.normal(0,3,500)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x1,y,s=1,label='x1',c='r')
plt.scatter(x2,y,s=1,label='x2',c='g')
plt.legend()
plt.show()
```

```python
X= np.array([x1,x2]).T # 2열의 2차원 배열이됨
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=0)
```

```python
model = Sequential()
model.add( Dense(2, input_dim=2))
model.add( Dense(64))
model.add( Dense(32))
model.add( Dense(16))
model.add( Dense(8))
model.add( Dense(4))
model.add( Dense(2))
model.add( Dense(1))
```

```python
model.compile(loss="mse", optimizer="adam", metrics=['mse'])
```

```python
hist = model.fit(X_train, y_train, epochs=500, batch_size=10,
                    validation_split=.1)
```

```python
hist.history.keys()
```

```python
loss = hist.history['loss']
mse = hist.history['mse']
val_loss = hist.history['val_loss']
val_mse = hist.history['val_mse']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,2))
plt.plot(loss, 'r--', label='loss')
plt.plot(mse, 'g', label='mse')
plt.plot(val_loss, 'y--', label='val_loss')
plt.plot(val_mse, 'b', label='val_mse')
plt.legend()
plt.show()
```

```python
model.evaluate(X_test, y_test)
```

```python
model.predict(np.array([[50,50]]))
```
