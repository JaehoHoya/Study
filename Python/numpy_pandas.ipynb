## ndarray 합치기(concatenate)
* 상하 방향으로 합치기(axis=0)
* 좌우 방향으로 합치기(axis=1)

```python
import numpy as np

arr1 = np.array(range(12)).reshape((3,4))
arr2 = np.array(range(12,20)).reshape((2,4))
arr1, arr2
```

```python
concat1 = np.concatenate((arr1, arr2))   # 디폴트 axis=0
print("Axis=0, 상하로 합친 결과:\n", concat1)
```

```python
# 축 지정(axis=1)
arr3 = np.array(range(20,26)).reshape((3,2))
concat2 = np.concatenate((arr1, arr3), axis=1)
print("Axis=1, 좌우로 합친 결과:\n", concat2)
```

## DataFrame CRUD
* df.loc[new index] = [한 행의 데이터]   # df에 새로운 한 행을 추가
* pd.concat([df, [새행 데이터]], axis=0, ignore_index=True)
* df.drop([인덱스,], axis=0)
* df.loc['a', 'price'] = 50

```python
# df_pet(name, color, weight, price, birth)
# df_pet 을 생성하고 한행을 추가하는 함수(pet_add()) 생성
```

```python
df_pet = pd.DataFrame(columns=['name', 'color', 'weight', 'price', 'birth'])
df_pet
```

```python
def pet_add():
    name = input('이름:')
    color = input('색상:')
    weight = float(input('몸무게:'))
    price = int(input('가격:'))
    birth = input('생일:')
    df_pet.loc[len(df_pet)] = [name, color, weight, price, birth]
    print('추가 완료')
```

```python
pet_add()
```

```python
df_pet
```

```python
# 특정 컬럼의 값을 사용하여 관련된 행의 인덱스 구하기
df_pet[df_pet['name']=='뽀메라이온'].index[0]    # 1
```

```python
# 이름(name)으로 검색하기
def find_by_name():
    name = input('이름:')
    found = df_pet[df_pet['name']==name]
    display(found)
```

```python
find_by_name()
```

```python
# 이름을 입력하여 weght, price 수정하기
# pet_update()
def pet_update():
    name = input('이름:')
    idx = df_pet[df_pet['name']==name].index[0]
    weight = float(input('몸무게:'))
    price = int(input('가격:'))
    df_pet.loc[idx, 'weight'] = weight
    df_pet.loc[idx, 'price'] = price
    print('수정 완료')
```

```python
pet_update()
```

```python
df_pet
```

```python
# 이름을 이용하여 행 삭제
def pet_delete():
    name = input('이름:')
    idx = df_pet[df_pet['name']==name].index[0]
    df_dropped = df_pet.drop(idx, axis=0)
    print('삭제 완료')
    display(df_dropped)
```

```python
pet_delete()
```

```python
df_pet
```

```python
df_pet
```

```python
# 컬럼 추가
# weight가 5보다 큰지 작은지 확인하여 작으면 0, 크면 1으로 설정(weight_label)
# weight_label 컬럼에 각 pet이 0인지 1인지를 표시하기
# df['weight_label'] = [행 수만큼의 데이터]
```

## apply(), map()
* apply() : DataFrame의 행이나 열을 대상으로 한 함수 적용(예, 최대/최소값 구하기)
* map() : Series에만 사용가능, 각 원소를 대상으로 함수 적용
* apply(), map()에는 주로 Lambda 식이 사용됨

```python
frame = DataFrame(np.random.randn(4, 3),
                  columns=['b', 'd', 'e'],
                  index=['Utah', 'Ohio', 'Texas', 'Oregon'])
frame
```

```python
np.abs(frame)    # 절대값
```

```python
f = lambda x: x.max() - x.min()   # Lambda Expression
```

```python
frame.apply(f)    # axis=0,
```

```python
frame.apply(f, axis=1)    # 각 행들의 최대/최소 차이값 구하기
```

```python
format = lambda x: '{:.2f}'.format(x)
```

```python
frame['e'].map(format)
```

```python
frame
```

```python
# 위에 정의된 frame의 각 컬럼의 평균을 화면에 표시해보세요.
avg = lambda x: x.mean()
```

```python
frame.apply(avg)    # axis=0
```

```python
frame.apply(avg, axis=1)    # axis=1
```

```python
# Series를 생성하고 임의의 값 5개를 채우고 각 원소들을 2로 나누기한 값을 표시한다
```

```python
ser = pd.Series(np.random.randn(5))
ser
```
