```python
# 사원정보 CRUD
# 번호, 이름, 전화번호를 목록에 저장(3인 정보)하고
# 키보드에서 전화번호를 이용하여 검색이 가능하게 함
emps = [
    {'no':11, 'name':'홍길동', 'phone':'010-2341-5630'},
    {'no':12, 'name':'홍길순', 'phone':'010-2341-5631'},
    {'no':13, 'name':'홍길자', 'phone':'010-2341-5632'}
]
```

```python
def find_by_phone(phone):
    for emp in emps:
        if emp['phone'] == phone:
            return emp
    return None
```

```python
def print_emp(emp):
    print(f"번호: {emp['no']}\t이름: {emp['name']}\t전화번호: {emp['phone']}")
```

```python
in_phone = input('전화번호 검색:')
emp = find_by_phone(in_phone)
if emp == None:
    print('검색된 사원이 없습니다.')
else:
    print(f"번호: {emp['no']}, 이름: {emp['name']}, 전화번호: {emp['phone']}")
```

```python
# 사원정보 검색, 추가, 수정, 삭제 함수 정의하고 사용하기
# 목록(s), 추가(a), 검색(f), 수정(u), 삭제(d), 종료(x)
```

```python
def emp_list():
    print("번호\t이름\t전화번호")
    for emp in emps:
        print(f"{emp['no']}\t{emp['name']}\t{emp['phone']}")
```

```python
def emp_add():
    no = int(input('번호:'))
    name = input('이름:')
    phone = input('전화번호:')
    emp = {'no':no, 'name':name, 'phone':phone}
    emps.append(emp)
    print('사원정보가 추가되었습니다.')
```

```python
def update_phone():
    new_phone = input('새 전화번호:')
    emp = find_by_phone(in_phone)
    if emp == None:
        print('검색된 사원이 없습니다.')
    else:
        emp['phone'] = new_phone
        print('사원정보가 수정되었습니다.')
```

```python
def delete_emp():
    empno = int(input('사원번호:'))
    for emp in emps:
        if emp['no'] == empno:
            emps.remove(emp)
            print('사원정보가 삭제되었습니다.')
            return
    print('검색된 사원이 없습니다.')
```

```python
while True:
    menu = input('목록(s), 추가(a), 검색(f), 수정(u), 삭제(d), 종료(x):')
    if menu == 's':
        emp_list()
    elif menu == 'a':
        emp_add()
    elif menu == 'f':
        in_phone = input('전화번호 검색:')
        print_emp(find_by_phone(in_phone))
    elif menu == 'u':
        update_phone()
    elif menu == 'd':
        delete_emp()
    elif menu == 'x':
        break
    else:
        print('메뉴입력 오류')
print('프로그램 종료')
```

```python
import random as rd

a = rd.randint(0,5)
b = rd.randint(0,5)

try:
    exp = f'{a} / {b} = {a/b}'
    print(exp)
except ZeroDivisionError:
    exp = f'{a} / {b} = 0으로 나눌 수 없습니다.'
    print(exp)
else:
    print('나눗셈에 성공했습니다.')
print('프로그램 종료')
```

```python
# 키보드에서 입력된 2개의 정수를 나눗셈하는 프로그램
# 이용자가 숫자가 아닌 문자를 입력한 경우의 반응을 확인
# 그 상황에서 예외처리를 해보세요.
import random as rd

try:
    a = int(input('정수입력:'))
    b = int(input('정수입력:'))

    exp = f'{a} / {b} = {a/b}'
    print(exp)
except ZeroDivisionError:
    exp = f'{a} / {b} = 0으로 나눌 수 없습니다.'
    print(exp)
except ValueError:
    print('숫자만 입력하세요')
else:
    print('나눗셈에 성공했습니다.')
print('프로그램 종료')
```

```python
# 키보드에서 입력된 2개의 정수를 나눗셈하는 프로그램
# 이용자가 숫자가 아닌 문자를 입력한 경우의 반응을 확인
# 그 상황에서 예외처리를 해보세요.
while True:
    try:
        a = int(input('정수입력:'))
        b = int(input('정수입력:'))

        exp = f'{a} / {b} = {a/b}'
        print(exp)
    except ZeroDivisionError as zde:
        exp = f'{a} / {b} = ' + str(zde)
        print(exp)
    except ValueError as ve:
        print('숫자만 입력하세요' + str(ve))
    else:
        print('나눗셈에 성공했습니다.')
        break
print('프로그램 종료')
```

```python
# 키보드에서 입력된 2개의 정수를 나눗셈하는 프로그램
# 이용자가 숫자가 아닌 문자를 입력한 경우의 반응을 확인
# 그 상황에서 예외처리를 해보세요.
while True:
    try:
        a = int(input('정수입력:'))
        b = int(input('정수입력:'))

        exp = f'{a} / {b} = {a/b}'
        print(exp)
    except Exception as ex:
        #print('예외발생:' + str(ex))
        if isinstance(ex, ZeroDivisionError):
            exp = f'{a} / {b} = ' + str(ex)
            print(exp)
        elif isinstance(ex, ValueError):
            print('숫자만 입력하세요')
    else:
        print('나눗셈에 성공했습니다.')
        break
print('프로그램 종료')
```

```python
# 수정/삭제
# 수정 : 길이, 체중 변경
# pet.txt 전체 데이터를 메모리에 로드
# 특정 pet을 이름으로 찾아서 검색된 pet의 길이, 체중 수정
# pet.txt 파일에 덮어쓰기
# pet 목록을 표시하여 제대로 수정되었는지 확인
```

```python
# pet.txt 파일을 읽어서 메모리에 로드한다
f = open('/content/drive/MyDrive/Python_Basic/pet.txt', mode='r', encoding='utf-8')
pets = f.readlines()
f.close()
print('pet.txt 로드됨')
```

```python
for i, pet in enumerate(pets):
    name, color, len, weight, age = pet.strip().split('|')
    if name == '풍산개':
        len = float(input('길이:'))
        weight = float(input('체중:'))
        pet = f'{name}|{color}|{len}|{weight}|{age}\n'
        pets[i] = pet
        print('메모리에서 수정됨')
        break
```

```python
f = open('/content/drive/MyDrive/Python_Basic/pet.txt', mode='w', encoding='utf-8')
for pet in pets:
    f.write(pet)
f.close()
print('pet.txt 수정됨')
```
